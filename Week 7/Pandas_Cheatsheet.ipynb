```python
def convert_for_temp(col):
    if col == 'NaN':
        return 38
    else:
        return col

def convert_for_day(col):
    if col == 'NaN':
        return 'Sunny'
    else:
        return col

def convert_for_windspeed(col):
    if col == 'NaN':
        return 40
    elif col=='Not available':
        return 35
    else:
        return col

season = pd.read_csv('data/season.csv', converters={
                      'day':convert_for_day,
                      'temp':convert_for_temp,
                      'wind-speed':convert_for_windspeed})
```

```python
# concat()

city_1 = {
    'date':['1-8-2022','2-8-2022','3-8-2022', '4-8-2022'],
    'event':['hot','cold','rainy','cold'],
    'temp':[28,13,25,17]
}

city_2 = {
    'date':['1-8-2022','2-8-2022','3-8-2022', '4-8-2022'],
    'event':['cold','windy','hot','rainy'],
    'temp':[12,21,29,22]
}

df_1 = pd.DataFrame(city_1)
df_2 = pd.DataFrame(city_2)

df = pd.concat([df_1,df_2], ignore_index=True)

df
```

```python
df = pd.concat([df_1,df_2], keys=['city_1','city_2'])

df
```

```python
df.loc['city_1','event']
```

```python
df = pd.concat([df_1,df_2], axis=1)
df
```

```python
# merge()
# merge is used to merge two dataframes together without repeating columns

df = pd.merge(df_1,df_2, on=['date'],suffixes=['_city_1','_city_2'])
df
```

```python
df = pd.merge(df_1,df_2, how='cross')
df
```

```python
# pivot_table() 
# allows you to reshape or trnasform your dataframe

dict_={
    'date':['5/1/2017', '5/2/2017','5/3/2017','5/1/2017', '5/2/2017','5/3/2017','5/1/2017', '5/2/2017','5/3/2017'],
    'city':['NYC','NYC','NYC','MUM','MUM','MUM','BNG','BNG','BNG'],
    'temp':[65,66,68,75,78,82,80,77,79],
    'humidity':[56,58,60,80,83,85,26,30,35]
}


df = pd.DataFrame(dict_)
df
```

```python
# To reshape the dataframe into another form you need to pass arguments, one is index='row'
# (row is the row which you want to be row of transformed dataframe) and other is columns='col'
# (col is the column which you want to be column of transformed dataframe)

df.pivot(index='date', columns='city')
```

```python
pd.pivot_table(df, index='city', columns='date', values=['temp'])
```

```python
df.pivot_table(index='date', columns='city',aggfunc='min')
```
