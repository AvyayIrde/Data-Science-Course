### Types of entity relationships 

#### One to one - Is used to create a relationship between 2 tables in which a single record of the first table can only be related to one and only one record of the second table, similarly the only record of the second table can be related to any one record of the first table. 

#### One to many - It is used to create a relationship between 2 tables, any single record of the first table can be related to one or more records of the second table but the records of the second table can only relate to only one record of the first table. 

#### many to many - it is used to to cretae a relationship between 2 tables, each record of the first table can relate to any records in the second table, likewise each record of the second table can also relate to more than one record of the first table.

### Data manipulation language 

#### DML modifies the database instance by inserting, updating and deleting its data. DML is responsible for all forms data modification in a database. SQL contains the following set of commands in its DML section −

#### SELECT/FROM/WHERE-
##### SELECT column_name(s) FROM table_name WHERE column operator value AND column operator value OR column operator value AND (... OR ...) 

#### INSERT INTO/VALUES -
##### INSERT INTO table_name VALUES (value_1, value_2,....)
##### INSERT INTO table_name (column1, column2,...) VALUES (value_1, value_2,....)

#### UPDATE/SET/WHERE-
##### UPDATE table_name SET column_name_1 = new_value_1, column_name_2 = new_value_2 WHERE column_name = some_value

#### DELETE FROM/WHERE-
##### DELETE FROM table_name WHERE column_name = some_value
