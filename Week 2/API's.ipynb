```python
# open-notify.com
#different endpoints poin tto different information

response = requests.get('http://api.open-notify.org/astros.json')
print(response.status_code)
print(response.content)

result = json.loads(response.content)
print(type(result))
```

```python
print("Total number of people in space rn = ", result['number'])
names = [name['name'] for name in result['people']]
print("\nNames of people in space rn :", names)
crafts = [craft['craft'] for craft in result['people']]
print("\nassociated crafts :", crafts)
```

```python
response = requests.get('http://api.open-notify.org/iss-now.json')
print(response.status_code)
print(response.text)

result = response.json() #instead of load

print(result.keys())
```

```python
print(f'Latitude and Longitide value', result['iss_position']['latitude'],result['iss_position']['longitude'])
time  = datetime.fromtimestamp(result['timestamp'])

print("current time of the data ", time)
```

```python
# http://api.open-notify.org/iss-pass.json

parameters = {
    "lat":12.4654,
    "lon":74.5135
}

url = 'http://api.open-notify.org/iss-pass.json'

response = requests.get(url=url,params=parameters)
# print(response.text)

result = response.json()

print(result)
```

```python
list_times = [datetime.fromtimestamp(time['risetime']) for time in result['response']]
for time in list_times:
    print(time)
    
```
