```python
import matplotlib.pyplot as plt
%matplotlib inline 
# produces charts inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46] # some frequency
```

```python
plt.rcParams['figure.figsize'] = (10, 6)
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature in the city')
plt.plot(x,y, color='red',linestyle='dotted',linewidth=2)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature in the city')
plt.plot(x,y,'b+--') # format string
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Temperature in the city')
plt.plot(x,y,color='b',marker='+',linestyle='--',markersize=15)
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='upper right',shadow=True,fontsize='medium')
```

```python
# bar chart
import numpy as np
```

```python
# to show the revenue of major tech companies
company=['Google','Apple','Facebook','Netflix']
revenue=[90,100,130,85]

plt.bar(company,revenue,label='Revenue')
```

```python
xpos = np.arange(len(company))
# print(xpos)
plt.bar(xpos,revenue,label='Revenue')
plt.xticks(xpos,company)
plt.ylabel('Revenue')
plt.title('Revenue of major tech companies')
plt.legend()
```

```python
# multiple bars showing revenue and profit
# vertical stacked bar chart

profit=[60,80,100,30]

plt.bar(company,revenue,label='Revenue',width=0.5)
plt.bar(company,profit,label='Profit',width=0.5)
plt.ylabel('Revenue')
plt.title('Revenue and Profit of major tech companies')
plt.legend()
```

```python
# side by side bar chart

plt.bar(xpos+0.2,revenue,label='Revenue',width=0.4)
plt.bar(xpos-0.2,profit,label='Profit',width=0.4)
plt.xticks(xpos,company)
plt.ylabel('Revenue')
plt.title('Revenue and Profit of major tech companies')
plt.legend()
```

```python
plt.barh(company,revenue,label='Revenue')
plt.barh(company,profit,label='Profit')
plt.xlabel('Revenue')
plt.ylabel('Company')
plt.title('Revenue and Profit of major tech companies')
plt.legend()
```

```python
# histogram

blood_sugar=[112,85,90,150,149,88,94,115,125,80,77,73,129]

plt.hist(blood_sugar,rwidth=0.5,bins=7,label='Blood Sugar')
```

```python
# 80-100 - normal
# 100-125 - pre diabetic
# >125 - diabetic

plt.xlabel('Blood Sugar')
plt.ylabel('Number of patients')
plt.title('Blood Sugar of patients')
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8,label='Blood Sugar', color='red')
```

```python
# multiple data samples in a histogram

plt.xlabel('Blood Sugar')
plt.ylabel('Number of patients')
plt.title('Blood Sugar of patients')

blood_sugar_men=[112,85,90,150,149,88,94,115,125,80,77,73,129]
blood_sugar_women=[67,98,99,120,133,150,84,69,1290,112,100]

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,label=['men','women'], color=['red','orange'])
plt.legend(loc='upper right',shadow=True,fontsize='medium')
```

```python
# histogram type

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8,label='Blood Sugar', color='red',histtype='step')
```

```python

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8,label='Blood Sugar', color='red',orientation='horizontal')
```

```python
# pie chart
# Track home expenses

exp_values=[100,200,300,400,500,600,700,800,900,1000]
exp_labels=['Others','Utilities','Food','Transport','Entertainment','Health','Debt','Insurance','Savings','Rent']

plt.pie(exp_values,labels=exp_labels) # perfect circle
plt.axis('equal')
plt.show()
```

```python
plt.pie(exp_values,labels=exp_labels,autopct='%1.1f%%',radius=1.5) # percentage
plt.axis('equal')
plt.savefig('pie_chart.jpeg',dpi=500)
plt.show()
```
